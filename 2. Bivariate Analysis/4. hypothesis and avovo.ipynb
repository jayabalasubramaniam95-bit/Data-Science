```python
#Does Gender and Work Experience affect MBA Percentage?
#mba_p -  Dependent Variable 
# Factor 1 - gender
# Factor 2 - work experiene
```

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm

# ols - Ordinary Least Squares
# It creates a linear regression model, which is also used internally to perform ANOVA.
# 'mba_p ~ C(gender) + C(workex) + C(gender):C(workex)' means Dependent Variable ~ Independent Variables
# C(gender) - statsmodels that gender is a categorical variable. Instead of treating values as numbers, it treats them as groups.
# C(gender) + C(workex) - Include both main effects in the model.
# C(gender):C(workex) - : means interaction effect.
#  Instead of testing gender and work experience separately, it asks, Does the effect of gender depend on work experience?
# .fit() - Calculate Regression coefficients Residuals (errors), Sum of Squares (SS), Mean Squares (MS), F-statistics Everything needed for the ANOVA table

model = ols(
    'mba_p ~ C(gender) + C(workex) + C(gender):C(workex)',
    data=dataset
).fit()

# anova_lm-  ANOVA for Linear Models.
# Calculate Sum of Squares (SS), Degrees of Freedom (df), Mean Square (MS), F-statistic p-value

sm.stats.anova_lm(model, typ=2)
```

```python
# PR(>F) - P values
# C(gender) - 0.000002 <0.05 - There is significant changes
# C(workex) - 0.002701 <0.05 - There is significant changes
# C(gender):C(workex) - 0.907393 > 0.05  - There is no significant changes
```

```python
#Does Work Experience and Specialisation affect Salary?
```

```python
model = ols(
    'salary ~ C(workex) + C(specialisation) + C(workex):C(specialisation)',
    data=dataset
).fit()

sm.stats.anova_lm(model, typ=2)
```

```python
# PR(>F) - P values
# C(workex) - 0.000132 <0.05 - There is significant changes
# C(specialisation) - 0.000525 <0.05 - There is significant changes
# C(workex):C(specialisation) - 0.640125 > 0.05  - There is no significant changes
```
